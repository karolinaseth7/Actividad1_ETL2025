By printing the first record of the dataset, we can observe the structure and format of each column. Some columns contain dictionaries within lists, others have nested dictionaries, while some store simple lists or standalone dictionaries. Due to this complexity, applying .dropna() directly is not feasible when handling null values, as it cannot extract data from within lists or dictionaries if they are present. The output below illustrates this issue, showing that certain columns contain null values.

- warranty                             60896
- seller_contact                       97781
- parent_item_id                       23011
- official_store_id                    99182
- differential_pricing                100000
- original_price                       99857
- video_id                             97015
- catalog_product_id                   99989
- subtitle                            100000

As seen in the first record above, the fields 'variations': [] and 'location': {} contain data, but they were not recognized by the isna() function. This happens because isna() only detects explicit null values (e.g., None or NaN) and does not consider non-null but empty structures like lists ([]) or dictionaries ({}). To properly determine whether these fields are empty, a different approach would be required, adding complexity to the process.

```python
null_counts = df.isnull().sum()
null_counts
```

The most complex structure shown above is the address column, which contains nested dictionaries. Additionally, some of these dictionaries hold multiple variables, making the structure even more intricate.

```python
df["seller_address"].iat[0]
```

```python
# Extract nested dictionaries and create a new column
df["country_name"] = df["seller_address"].apply(lambda x: x["country"]["name"] if isinstance(x, dict) and "country" in x else None)
df["country_id"] = df["seller_address"].apply(lambda x: x["country"]["id"] if isinstance(x, dict) and "id" in x else None)
df["shipping_local_pick_up"] = df["shipping"].apply(lambda x: x.get("local_pick_up") if isinstance(x, dict) else None)
df["shipping_free_shipping"] = df["shipping"].apply(lambda x: x.get("free_shipping") if isinstance(x, dict) else None)
df["shipping_mode"] = df["shipping"].apply(lambda x: x.get("mode") if isinstance(x, dict) else None)

df
```

The easiest way to check for null or NaN values in nested dictionaries is to create new columns for their data. This makes it simpler to analyze and apply the necessary checks for each case.

```python
conteo_1 = df["country_name"].value_counts()
conteo_1
```

```python
num_country = df["country_name"].isnull().sum()
num_country
```

```python
unique_countries = df["country_name"].unique()
unique_countries
```

```python
#convert the data_created to a datetime type
df["date_created"] = pd.to_datetime(df["date_created"])
```

## NEW TABLE

```python
data_select = ["id", "condition" , "title", "price", "date_created", "geolocation" ,
                "status", "permalink", "available_quantity", "shipping_local_pick_up"
                , "shipping_free_shipping", "shipping_mode"]
df_selected = df[data_select]
df_selected
```

```python
df_selected.isnull().sum()
```

### NORMALIZATION

```python
df_selected["latitude"] = df_selected["geolocation"].apply(lambda x: x.get("latitude") if isinstance(x, dict) else None)
df_selected["longitude"] = df_selected["geolocation"].apply(lambda x: x.get("longitude") if isinstance(x, dict) else None)
df_selected["año"] = df_selected["date_created"].dt.year
df_selected["mes"] = df_selected["date_created"].dt.month
df_selected["día"] = df_selected["date_created"].dt.day
df_selected["día_semana"] = df_selected["date_created"].dt.day_name()
df_selected

```
